### IMPORTING LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

###DATA INGESTION 

```python
data_path="/Workspace/Users/terrysenosi@gmail.com/Python-boot-camp-Exercises/data source/retail_sales_dataset (2).csv"
```

```python
df=pd.read_csv(data_path)
df.head(5)
```

#1.EXERCISE ONE

```python
# Converting date column to datetime
df["date"] = pd.to_datetime(df["Date"])
# Extracting month from date column
df["month"] = df["date"].dt.month_name()

#creating figure
fig,ax = plt.subplots(2,2,figsize=(14,19))

#plotting graph1, bar chart, revenue by product category
revenue_by_product = df.groupby("Product Category")["Total Amount"].sum()
revenue_by_product.plot.bar(ax=ax[0,0])
ax[0,0].set_title("Revenue by product category")
ax[0,0].set_xlabel("Product Category")
ax[0,0].set_ylabel("Total Amount")
ax[0,0].tick_params(axis='x', rotation=45)

#plotting graph 2, line chart, total revenue peer month
total_revenue_per_month = df.groupby("month")["Total Amount"].sum()
total_revenue_per_month.plot.line(ax=ax[0,1])
ax[0,1].set_title("Revenue per month")
ax[0,1].set_xlabel("Month")
ax[0,1].set_ylabel("Total Amount")
, rotation=45

#plotting graph 3, histogram, distribution of revenue
df["Total Amount"].plot.hist(ax=ax[1,0])
ax[1,0].set_title("Distribution of revenue")
ax[1,0].set_xlabel("Total Amount")
ax[1,0].set_ylabel("Frequency")

#plotting graph 4, multiline chart, revenue by product category per month
revenue_by_product_per_month = df.groupby(["month","Product Category"])["Total Amount"].sum()
revenue_by_product_per_month.unstack().plot.line(ax=ax[1,1])
ax[1,1].set_title("Revenue by product category per month")
ax[1,1].set_xlabel("Month")
ax[1,1].set_ylabel("Total Amount")
ax[1,1].legend()

```

```python
#plotting graph 4, multiline chart, revenue by product category per month

revenue_by_product_per_month = df.groupby(["Date","Product Category"])["Total Amount"].sum()
revenue_by_product_per_month.unstack().plot.line()
plt.figure(figsize=(10,5))
plt.title("Revenue by product category per month")
plt.xlabel("Date")
plt.ylabel("Total Amount")
plt.xticks(rotation=45) #
plt.tight_layout #makes layout neat
plt.show()
```
