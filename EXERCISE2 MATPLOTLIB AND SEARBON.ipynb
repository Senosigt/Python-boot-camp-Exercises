```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
```

```python
data_path="/Workspace/Users/terrysenosi@gmail.com/Python-boot-camp-Exercises/data source/retail_sales_dataset (2).csv"
```

```python
df=pd.read_csv(data_path)
df.head(5)
```

```python
df.columns
```

```python
# Converting date column to datetime
df["date"] = pd.to_datetime(df["Date"])
# Extracting month and day of week from date column
df["month"] = df["date"].dt.month_name()
df["day_of_week"] = df["date"].dt.day_name()

#creating figure
fig,ax = plt.subplots(1,2,figsize=(12,8))

#plotting graph1 on left, boxplot of total amount by product category
sns.boxplot(x="Product Category",y="Total Amount",data=df,ax=ax[0])
ax[0].set_title("Total Amount by Product Category")
ax[0].set_xlabel("Product Category")
ax[0].set_ylabel("Total Amount")
ax[0].tick_params(axis='x', rotation=90)

#plotting graph2 on right,histogram of total amount for the highest variance products
high_var_product=df.groupby("Product Category")["Total Amount"].std().sort_values(ascending=False).index[0]
sns.histplot(x="Total Amount",data=df[df["Product Category"]==high_var_product],ax=ax[1])
ax[1].set_title("Total Amount for Highest Variance Product")
ax[1].set_xlabel("Total Amount")
ax[1].set_ylabel("Frequency")
plt.show()
```
